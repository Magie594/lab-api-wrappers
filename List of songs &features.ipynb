```python
from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3000)/1000) # respectful nap
    return tracks
```

```python
all_tracks = get_playlist_tracks("4rnleEAOdmFAbRcNCgZMpY")
len(all_tracks)
```

```python
import pandas as pd
from pandas import json_normalize

tracks2= json_normalize(all_tracks)
```

```python
# probably let them figure this out in the lab :-)
chunks = [(i, i+100) for i in range(0, len(tracks2), 100)]
audio_features_list = []
for chunk in chunks:
    id_list100 = tracks2['track.id'][chunk[0]:chunk[1]]
    audio_features_list = audio_features_list + sp.audio_features(id_list100)
    sleep(randint(1,3000)/1000)
len(audio_features_list)
```

```python
import pandas as pd
from pandas import json_normalize

audio_features_df = json_normalize(audio_features_list)
```

```python
audio_features_df
```

```python
audio_features_df = audio_features_df.drop_duplicates()
```

```python
df_w_audio_ft = pd.merge(left=tracks2,
                        right=audio_features_df,
                        how='inner',
                        left_on='track.uri',
                        right_on='uri')
df_w_audio_ft
```
